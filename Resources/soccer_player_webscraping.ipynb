```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
result = []

r = requests.get("https://www.transfermarkt.co.uk/fc-chelsea/startseite/verein/631/saison_id/2020", headers= {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:74.0) Gecko/20100101 Firefox/74.0"})
soup = BeautifulSoup(r.content, "html.parser")

tr = soup.find_all("tbody")[1].find_all("tr", recursive=False)

result.extend([
        { 
            "Player": t[1].find_all("a", {"class": "spielprofil_tooltip"})[0].text,
            "Club": t[4].find("img")["alt"],
            "Nationality": t[3].find("img")["alt"] 
        }
        for t in (t.find_all(recursive=False) for t in tr)
    ])

df = pd.DataFrame(result)
```

```python
df
```

```python
team = df.loc[df["Club"] == "Chelsea FC"]
team
```

```python
# Export to CSV
team.to_csv("chelsea_players.csv", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

page = "https://www.transfermarkt.co.uk/manchester-united/startseite/verein/985?saison_id=2020"
pageTree = requests.get(page, headers=headers)
soup = BeautifulSoup(pageTree.content, 'html.parser')
```

```python
results = soup.find_all("span", class_ = "hide-for-small")
results
for result in results:
    name = result.find("a", class_="spielprofil_tooltip")
    try:
        print(name.text)
    except:
        print("")
```

```python
players = soup.find_all("a", {"class": "spielprofil_tooltip"})
for player in players:
    print(player.text)
```

```python
imgs = [a.find('img') for a in soup.find_all("td", {"class": "zentriert"}) if a.find('img')]
nationality = [img.get('alt') for img in imgs]
nationality
```
